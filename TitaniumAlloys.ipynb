#This is a notebook about titanium and its alloys

Titanium is a silvery white metal that has a hexagonal close packed structure at room temperature but this changes to body centred cubic at higher temperatures. The phase change is important in alloys because as in steels the phase change can be used to alter the mechanical properties of the alloy and a martensitic type of structure is possible. Alloys are mainly based on Titanium with aluminium and vanadium. A typical phase diagram for a titanium alloy is shown below. (Source CES Edupack 2019).
