Importing an image by calling matplotlib.pyplot

Example from my drive

```
import matplotlib.pyplot as plt
import matplotlib.image as img
%matplotlib inline
image = img.imread("/content/drive/My Drive/Titanium/titanium plates.png")
plt.imshow(image)
plt.show()
```

As above but with a couple of extra lines so you can change the image shape.

```
import matplotlib.pyplot as plt
import matplotlib.image as img
%matplotlib inline
plt.rcParams['figure.figsize'] = [8, 5]
image = img.imread("/content/drive/My Drive/Titanium/titanium plates.png")
plt.axis('off')
plt.imshow(image)
plt.show()
```

```
import matplotlib.pyplot as plt
import matplotlib.image as img
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 8]
image = img.imread("/content/drive/My Drive/Titanium/ttt_ti.jpg")
plt.axis('off')
plt.imshow(image)
plt.show()
```
