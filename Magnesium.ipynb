#Magnesium and its alloys

Magnesium is a light material (density 1738 Mg/m3) in CES Edupack on level 3 
there are 40 alloys listed. Of the three 'light' alloys in the aircraft 
industry they are the least dense as the figure below shows.

# Phase Diagrams

The phase diagram below shows the Mg - Al system. Most alloys contain less than 10 wt% Aluminium. You can see from the diagram below that Al can dissolve up to 12% aluminium at the eutectic temperature. The primary phase is a magnesium rich solid solution.
